### Zadanie 1 — *budowa macierzy + przekątna*
Zbuduj macierz **6×6** wypełnioną liczbami `1..36`, a następnie wyciągnij jej **główną przekątną**.

<details><summary>💡 Podpowiedź</summary>

### Zadanie 2 — *broadcasting tablica + wektor*
Mając macierz `A = np.arange(12).reshape(4, 3)`, dodaj do **każdego wiersza** wektor `np.array([100, 200, 300])`.

<details><summary>💡 Podpowiedź</summary>

Wektor o kształcie `(3,)` „rozlewa się" (broadcasting) na każdy wiersz macierzy `(4, 3)` — wystarczy zwykłe `A + wektor`.
</details>

<details><summary>✅ Rozwiązanie</summary>

```python
A = np.arange(12).reshape(4, 3)
A + np.array([100, 200, 300])
```
</details>

### Zadanie 3 — *normalizacja kolumn (z-score)*
Dla macierzy `X = np.array([[10, 1], [20, 3], [30, 5], [40, 7]])` znormalizuj **każdą kolumnę osobno**: od każdej wartości odejmij średnią jej kolumny i podziel przez odchylenie standardowe tej kolumny.

<details><summary>💡 Podpowiedź</summary>

Policz `X.mean(axis=0)` i `X.std(axis=0)` — oba mają kształt `(2,)`, więc dzięki broadcastingowi `(X - mean) / std` zadziała na całą macierz.
</details>

<details><summary>✅ Rozwiązanie</summary>

```python
X = np.array([[10, 1], [20, 3], [30, 5], [40, 7]])
(X - X.mean(axis=0)) / X.std(axis=0)
```
</details>

### Zadanie 4 — *warunkowe podstawianie (np.where)*
Mając tablicę `t = np.array([-5, 3, -1, 8, 0, -7, 2])`, utwórz nową tablicę, w której **wartości ujemne zamienisz na 0**, a pozostałe zostawisz bez zmian (to funkcja ReLU).

<details><summary>💡 Podpowiedź</summary>

`np.where(warunek, wartość_gdy_prawda, wartość_gdy_fałsz)`.
</details>

<details><summary>✅ Rozwiązanie</summary>

```python
t = np.array([-5, 3, -1, 8, 0, -7, 2])
np.where(t < 0, 0, t)
```
</details>

### Zadanie 5 — *agregacja + argmax*
Dla macierzy `M = np.array([[3, 8, 1], [9, 2, 5], [4, 4, 4]])` policz **sumę każdego wiersza**, a następnie wskaż **indeks wiersza o największej sumie**.

<details><summary>💡 Podpowiedź</summary>

`M.sum(axis=1)` da sumy wierszy, a `.argmax()` zwróci indeks największej z nich.
</details>

<details><summary>✅ Rozwiązanie</summary>

```python
M = np.array([[3, 8, 1], [9, 2, 5], [4, 4, 4]])
sumy = M.sum(axis=1)
print(sumy)
sumy.argmax()
```
</details>

### Zadanie 6 — *sortowanie wierszy wg kolumny*
Mając macierz `dane = np.array([[3, 30], [1, 10], [2, 20]])`, posortuj **całe wiersze** rosnąco według wartości w **pierwszej kolumnie**.

<details><summary>💡 Podpowiedź</summary>

`np.argsort` na pierwszej kolumnie (`dane[:, 0]`) zwróci kolejność wierszy — użyj jej do przeindeksowania: `dane[kolejność]`.
</details>

<details><summary>✅ Rozwiązanie</summary>

```python
dane = np.array([[3, 30], [1, 10], [2, 20]])
kolejnosc = np.argsort(dane[:, 0])
dane[kolejnosc]
```
</details>

### Zadanie 7 — *iloczyn macierzy*
Pomnóż macierzowo `A = np.arange(6).reshape(2, 3)` przez `B = np.arange(6).reshape(3, 2)`. Jaki kształt ma wynik?

<details><summary>💡 Podpowiedź</summary>

Iloczyn macierzy to operator `@` (lub `np.dot`). `(2,3) @ (3,2)` → `(2,2)`. To **nie** to samo co `*` (mnożenie element po elemencie)!
</details>

<details><summary>✅ Rozwiązanie</summary>

```python
A = np.arange(6).reshape(2, 3)
B = np.arange(6).reshape(3, 2)
A @ B
```
</details>

### Zadanie 8 — *maska 2D + zliczanie*
Dla macierzy `M = np.arange(1, 26).reshape(5, 5)` policz, **ile elementów jest większych od średniej całej macierzy**.

<details><summary>💡 Podpowiedź</summary>

Zbuduj maskę `M > M.mean()` (tablica `True`/`False`), a liczbę `True` policzysz przez `.sum()` lub `np.count_nonzero(...)`.
</details>

<details><summary>✅ Rozwiązanie</summary>

```python
M = np.arange(1, 26).reshape(5, 5)
(M > M.mean()).sum()
```
</details>

### Zadanie 9 — *(dla chętnych) tabliczka mnożenia bez pętli*
Zbuduj macierz **10×10** będącą tabliczką mnożenia (element w wierszu `i`, kolumnie `j` to `i*j` dla `i, j` od 1 do 10) — **bez żadnej pętli `for`**, samym broadcastingiem.

<details><summary>💡 Podpowiedź</summary>

Weź wektor pionowy `(10, 1)` i poziomy `(1, 10)` i pomnóż je przez siebie — broadcasting zrobi resztę. Wektor pionowy uzyskasz przez `reshape(-1, 1)`.
</details>

<details><summary>✅ Rozwiązanie</summary>

```python
v = np.arange(1, 11)
v.reshape(-1, 1) * v        # (10,1) * (10,) -> (10,10)
```
</details>
