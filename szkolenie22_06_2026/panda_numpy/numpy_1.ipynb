## 1. 📥 Tworzenie tablic

Bibliotekę importujemy konwencjonalnie jako `np`. Tablicę zrobimy z listy (`np.array`) albo jedną z funkcji generujących:

| Funkcja | Co robi |
|---------|---------|
| `np.array([...])` | tablica z istniejącej listy |
| `np.arange(start, stop, krok)` | kolejne liczby (jak `range`) |
| `np.zeros(n)` / `np.ones(n)` | same zera / jedynki |
| `np.linspace(a, b, ile)` | `ile` równo rozłożonych liczb od `a` do `b` |

````python
import numpy as np

a = np.array([1, 2, 3, 4, 5])
a
````

````python
# Kolejne liczby 0..9
np.arange(10)
````

````python
# 5 liczb równo rozłożonych od 0 do 1
np.linspace(0, 1, 5)
````

````python
# Tablica dwuwymiarowa (macierz 2x3) z zagnieżdżonej listy
m = np.array([[1, 2, 3],
              [4, 5, 6]])
m
````

<a id="2"></a>
## 2. 🔤 Atrybuty tablicy

Każda tablica „zna" swój kształt i typ. Najważniejsze atrybuty:

| Atrybut | Znaczenie |
|---------|-----------|
| `shape` | kształt, np. `(2, 3)` = 2 wiersze × 3 kolumny |
| `ndim`  | liczba wymiarów |
| `size`  | łączna liczba elementów |
| `dtype` | typ danych (np. `int64`, `float64`) |

````python
print('shape:', m.shape)
print('ndim: ', m.ndim)
print('size: ', m.size)
print('dtype:', m.dtype)
````

<a id="3"></a>
## 3. 🎯 Indeksowanie i wycinki

Indeksujemy jak listy (od `0`), ale w 2D podajemy **wiersz i kolumnę naraz**: `m[wiersz, kolumna]`.

Wycinki (`slice`) działają jak w listach — `start:stop` (stop wyłączony), a `:` oznacza „wszystko" w danym wymiarze.

````python
# Pojedynczy element tablicy 1D
a[0]
````

````python
# Wycinek 1D — elementy o indeksach 1, 2, 3
a[1:4]
````

````python
# Element z macierzy: wiersz 0, kolumna 2
m[0, 2]
````

````python
# Cała druga kolumna (wszystkie wiersze, kolumna o indeksie 1)
m[:, 1]
````

<a id="4"></a>
## 4. ➕ Operacje wektorowe (broadcasting)

Działania wykonują się **na wszystkich elementach naraz**, bez pętli. Dodanie liczby do tablicy zadziała na każdy element (to właśnie *broadcasting*), a dwie tablice tej samej długości łączymy element po elemencie.

````python
# Każdy element pomnożony przez 10
a * 10
````

````python
# Działanie element po elemencie między dwiema tablicami
b = np.array([10, 20, 30, 40, 50])
a + b
````

````python
# Funkcje też działają na całej tablicy naraz
np.sqrt(a)
````

<a id="5"></a>
## 5. 📊 Agregacje i statystyki

NumPy ma wbudowane metody podsumowujące całą tablicę: `sum`, `mean`, `min`, `max`, `std` (odchylenie standardowe) i inne.

💡 W tablicach 2D parametr **`axis`** mówi, *wzdłuż którego wymiaru* liczyć:
- `axis=0` → w dół, **po kolumnach**
- `axis=1` → w bok, **po wierszach**

````python
print('suma:    ', a.sum())
print('średnia: ', a.mean())
print('min/max: ', a.min(), a.max())
print('odch.std:', a.std())
````

````python
# Suma po kolumnach (axis=0) — jedna liczba na kolumnę
m.sum(axis=0)
````

````python
# Suma po wierszach (axis=1) — jedna liczba na wiersz
m.sum(axis=1)
````

<a id="6"></a>
## 6. 🔎 Filtrowanie maską logiczną

Porównanie na tablicy zwraca **maskę** wartości `True`/`False`. Wstawiając tę maskę w nawiasy kwadratowe, wyciągamy tylko pasujące elementy.

**Łączenie warunków** (jak w pandas — operatory bitowe, nie `and`/`or`!):

| Operator | Znaczenie |
|----------|-----------|
| `&` | **i** (AND) |
| `\|` | **lub** (OR) |
| `~` | **negacja** |

⚠️ Każdy warunek w osobnych nawiasach: `a[(warunek1) & (warunek2)]`.

````python
# Sama maska — które elementy są > 2?
a > 2
````

````python
# Wyciągamy tylko elementy spełniające warunek
a[a > 2]
````

````python
# Dwa warunki naraz — elementy większe od 1 ORAZ mniejsze od 5
a[(a > 1) & (a < 5)]
````

<a id="7"></a>
## 7. 🔄 Zmiana kształtu

Te same dane można „przeformować" w inny kształt:

| Metoda | Co robi |
|--------|---------|
| `reshape(w, k)` | nadaje nowy kształt (iloczyn musi się zgadzać z `size`) |
| `ravel()` | spłaszcza do 1D |
| `.T` | transpozycja (zamiana wierszy z kolumnami) |

💡 W `reshape` można podać `-1` w jednym wymiarze — NumPy sam policzy jego rozmiar.

````python
# 12 kolejnych liczb ułożone w macierz 3x4
siatka = np.arange(12).reshape(3, 4)
siatka
````

````python
# Transpozycja — wiersze stają się kolumnami
siatka.T
````

````python
# Spłaszczenie z powrotem do jednego wymiaru
siatka.ravel()
````

<a id="8"></a>
## 8. 🐍 Powrót do Pythona

Tablicę NumPy zamieniamy z powrotem na **zwykłą listę Pythona** metodą `tolist()` (tablica 2D da listę list). Pojedynczy element wyciągnięty z tablicy zamienimy na natywny typ przez `item()`.

````python
# Tablica 1D -> lista
a.tolist()
````

````python
# Tablica 2D -> lista list
lista = m.tolist()
print(type(lista), type(lista[0]))
lista
````
